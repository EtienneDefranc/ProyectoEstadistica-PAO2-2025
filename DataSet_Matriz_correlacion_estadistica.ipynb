# Task
Analyze feature correlations in the `EV_Energy_Consumption_Cleaned.csv` dataset, visualize them using a heatmap, and specifically examine the correlation between 'Temperature_C' and 'Battery_Temperature_C' to recommend whether to keep both or only one to avoid multicollinearity if their correlation is above 0.8.

## Load Data

### Subtask:
Load the cleaned data from the 'EV_Energy_Consumption_Cleaned.csv' file into a pandas DataFrame.


**Reasoning**:
To load the data, I need to import the pandas library, read the CSV file into a DataFrame, and then display the first 5 rows to verify the loading process.



```python
import pandas as pd

df = pd.read_csv('/content/EV_Energy_Consumption_Cleaned.csv')
print(df.head())
```

## Generate Correlation Matrix

### Subtask:
Calculate the correlation matrix for all numerical columns in the DataFrame.


**Reasoning**:
To calculate the correlation matrix, I will use the .corr() method on the DataFrame, which automatically computes pairwise correlations for all numerical columns.



```python
correlation_matrix = df.corr()
print(correlation_matrix.head())
```

**Reasoning**:
To visualize the correlations, I will generate a heatmap of the correlation matrix. This requires importing `matplotlib.pyplot` and `seaborn` for plotting.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of EV Energy Consumption Features')
plt.show()
```

**Reasoning**:
To specifically examine the correlation between 'Temperature_C' and 'Battery_Temperature_C', I will extract their correlation coefficient from the already computed correlation matrix. Then, I will apply the given condition to decide if both should be kept.



```python
temp_battery_temp_corr = correlation_matrix.loc['Temperature_C', 'Battery_Temperature_C']

print(f"Correlation between Temperature_C and Battery_Temperature_C: {temp_battery_temp_corr:.2f}")

if abs(temp_battery_temp_corr) > 0.8:
    print("Recommendation: The correlation between 'Temperature_C' and 'Battery_Temperature_C' is high (above 0.8). Consider keeping only one of these features to avoid multicollinearity.")
else:
    print("Recommendation: The correlation between 'Temperature_C' and 'Battery_Temperature_C' is not exceptionally high (below 0.8). It is acceptable to keep both features.")
```

## Final Task

### Subtask:
Summarize the findings from the correlation matrix, specifically advising on whether to keep both 'Temperature_C' and 'Battery_Temperature_C' or choose only one if their correlation is above 0.8, to avoid multicollinearity.


## Summary:

### Q&A
The correlation between 'Temperature_C' and 'Battery_Temperature_C' is -0.04. Since this value is below the 0.8 threshold, it is recommended to keep both 'Temperature_C' and 'Battery_Temperature_C' as multicollinearity is not a significant concern between these two features.

### Data Analysis Key Findings
